```python
import tweepy
import os
import pandas as pd
import numpy as np
consumer_key = ''
consumer_secret = ''
access_token = ''
access_token_secret = ''

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth,wait_on_rate_limit=True)
```

```python
print(api.search(q='#JoeBiden', lang = 'en', count = 10, since = "2021-01-25"))
```

```python
i = 1
joe = tweepy.Cursor(api.search,q='#JoeBiden', lang = 'en', since = '2021-01-25').items(1000)
for tweet in joe:
    print(str(i),') ',tweet.created_at, ':', tweet.text,'\n')
    details = [[tweet.created_at, tweet.text] for tweet in joe]
    df = pd.DataFrame(data = details, columns = ['Date', 'Tweets'])
    #print(df)
    i = i+1
    
print(df)
```

```python
df100=(df[-100:])
print(df100)
```
